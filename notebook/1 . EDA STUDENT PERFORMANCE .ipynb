### 3.8 Adding columns for "Total Score" and "Average"

```python
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average_score'] = df['total_score'] / 3

df.head()
```

# Students with full scores across all subjects

```python
# Count the number of students with a math score of 100
math_full_score_count = len(df[df['math_score'] == 100])

# Count the number of students with a reading score of 100
reading_full_score_count = len(df[df['reading_score'] == 100])

# Count the number of students with a writing score of 100
writing_full_score_count = len(df[df['writing_score'] == 100])

# Display the count
print("Number of students with full score in math:", math_full_score_count)

print("Number of students with full score in reading:", reading_full_score_count)

print("Number of students with full score in writing:", writing_full_score_count)
```

# Students who secured 20 or less accross all Subjects

```python
#Lets take a look at the students who scored below 20 
#in each of the subjects

math_below_20 = (df['math_score'] <= 20).sum()
print(f"Students who secured 20 or below in Math: {math_below_20}")

reading_below_20 = (df['reading_score'] <= 20).sum()
print(f"Students who secured 20 or below in Reading: {reading_below_20}")

writing_below_20 = (df['writing_score'] <= 20).sum()
print(f"Students who secured 20 or below in Writing: {writing_below_20}")
```

#####  Insights
 - From above values we get that students have performed the worst in Maths 
 - Best performance is in reading
